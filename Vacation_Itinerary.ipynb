```python
# Import dependencies.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import gmaps
import requests

from config import g_key
```

```python
selected_cities = pd.read_csv("data/WeatherPy_vacation.csv")
selected_cities
```

```python
my_itenary =  selected_cities[selected_cities["Country"] == "BR"]
my_itenary
```

```python
my_itenary.reset_index(drop=True,inplace=True)
my_itenary
```

```python
city_a = (my_itenary.loc[0]["Lat"],my_itenary.loc[0]["Lng"])
city_b = (my_itenary.loc[1]["Lat"],my_itenary.loc[1]["Lng"])
city_c = (my_itenary.loc[2]["Lat"],my_itenary.loc[2]["Lng"])
city_d = (my_itenary.loc[6]["Lat"],my_itenary.loc[6]["Lng"])

city_a
city_b
city_c
city_d
```

```python
selectedItenary_df = my_itenary[(my_itenary["City"] == "Malacacheta") | (my_itenary["City"] == "Rio Verde De Mato Grosso") | (my_itenary["City"] == "Jatai") | (my_itenary["City"] == "Jales")]  
# selectedItenary_df.append(my_itenary.loc[4:6])
selectedItenary_df
```

```python
import gmaps.datasets

gmaps.configure(api_key=g_key)

fig = gmaps.figure()

city_a2city_d = gmaps.directions_layer(city_a, city_d, waypoints=[city_b, city_c],travel_mode='DRIVING')

fig.add_layer(city_a2city_d)
fig
```

```python
import gmaps

gmaps.configure(api_key=g_key)

fig = gmaps.figure()

#Adding markers.
info_box_template = """
<dl>
<dt>Hotel Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
<dt>Current Weather</dt><dd>{Current Description} and {Max Temp} Â°F</dd>
</dl>
"""
# Store the DataFrame Row.
hotel_info = [info_box_template.format(**row) for index, row in selectedItenary_df.iterrows()]

#Create the marker for each city.
locations = selectedItenary_df[["Lat","Lng"]]
marker_layer = gmaps.marker_layer(locations, info_box_content=hotel_info)
fig.add_layer(marker_layer)

#direction Layer
city_a2city_d = gmaps.directions_layer(city_a, city_d, waypoints=[city_b, city_c],travel_mode='DRIVING')
fig.add_layer(city_a2city_d)
fig
```
